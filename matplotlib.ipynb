```python
import matplotlib.pyplot as plt

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

product_a = [1200, 1350, 1420, 1600, 1700, 1550, 1480, 1600, 1650, 1750, 1900, 2000]
product_b = [980, 1100, 1150, 1170, 1250, 1300, 1250, 1280, 1340, 1400, 1500, 1600]
product_c = [750, 770, 800, 850, 900, 950, 970, 990, 1020, 1100, 1200, 1250]
marketing_spend = [300, 320, 400, 500, 550, 600, 580, 620, 650, 670, 720, 750]

total_sales = [a + b + c for a, b, c in zip(product_a, product_b, product_c)]
print(total_sales)

```

```python
# Creating Line Plot of Above Data
plt.figure(figsize=(16, 6))
plt.plot(months, product_a, label='Product A', marker = 'o')
plt.plot(months, product_b, label='Product B', marker = '*')
plt.plot(months, product_c, label='Product C', marker = 'x')
plt.title('Monthly Sales of Products')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Comparision of Products Sales in Specific months
```

```python
import numpy as np

plt.figure(figsize=(10, 6))
bar_width = 0.25
x = np.arange(len(months))
print (x)

plt.bar(x-0.25, product_a, width=bar_width, label='Product A')
plt.bar(x, product_b, width=bar_width, label='Product B')
plt.bar(x+0.25, product_c, width=bar_width, label='Product C')

plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Product Sales Comparison')
plt.xticks(x, months)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(months, product_a, label='Product A')
plt.bar(months, product_b, bottom=product_a, label='Product B')
bottom_ab = [a + b for a, b in zip(product_a, product_b)]
plt.bar(months, product_c, bottom=bottom_ab, label='Product C')

plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Stacked Sales by Product')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot([product_a, product_b, product_c], labels=['Product A', 'Product B', 'Product C'])
plt.title('Distribution of Monthly Sales')
plt.ylabel('Sales')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(marketing_spend, total_sales, c='green', s=100)
plt.title('Marketing Spend vs. Total Sales')
plt.xlabel('Marketing Spend ($)')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
