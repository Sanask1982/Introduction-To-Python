**Ploting Data in Different Styles**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def Styles():


  style="fivethirtyeight" # @param ["grayscale", "dark_background", "fivethirtyeight", "ggplot", "Solarize_Light2"] {"allow-input":true, isTemplate: true}
  print("Your Style is ", style)

  data=pd.Series(np.random.randn(1000).cumsum())  #cumulative sum
  plt.style.use(style)
  data.plot()
  plt.show()
Styles()
```
