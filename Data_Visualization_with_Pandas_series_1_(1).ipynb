# Data Visualization with Pandas - 1

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

** Plot style is dark backgroud**

```python
plt.style.use("dark_background")
```

## Plotting graph with plot() method

**Generating Data Using random.rand(1000) which means 0 to 999 rows and "data" is storing variable**

```python
data=pd.Series(np.random.randn(1000).cumsum())  #cumulative sum
data
```

```python
data.plot()
```

**Generating random data using numpy function random.randn(100,4) which means random numbers in 100 rows and 4 columns and "df1" is variable

```python
df1 = pd.DataFrame(np.random.randn(100, 4),columns=list('ABCD'))
df1
```

** Ploting cumulative sum of stored data in df1 **

```python
df1 = df1.cumsum() # cumsum() is  cumulative sum
```

```python
df1.plot()
```

## Bar Charts Examples






**iloc[10]selects data from a DataFrame by row and column number from "df1" and plotting that data in bar graph**

```python

df1.iloc[10].plot(kind='bar')
```

** generating data using random.rand(7,3) means 7 rows and 3 columns**

```python
df2=pd.DataFrame(np.random.rand(7,3), columns=list("ABC")) # Column names
df2
```

**generating bar graph **

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

**horizontal bar chart**

```python
df2.plot.barh(stacked=True)
```

## Histograms

**Pulling Data from Github Directly**

```python
csv_url = "https://github.com/plotly/datasets/blob/master/iris-data.csv?raw=true"
iris = pd.read_csv(csv_url)
```

**Using this data**

```python

iris.columns=["sepal_length","sepal_width", "petal_length",
              "petal_width", "species"]
```

You can access this data set [here](https://github.com/tirendazacademy/data-visualization-with-python).

```python
iris.head()
```

```python
iris.plot.hist(alpha=0.7)
```

```python
iris.plot.hist(alpha=1, stacked=True)
```

```python
bins=25
```

```python
iris.plot.hist(alpha=1, stacked=True, bins=25)
```

```python
iris["sepal_width"].plot.hist(orientation="horizontal")
```

```python
iris["sepal_length"].diff().hist()
```

```python
iris.hist(color="blue", alpha=1, bins=20)
```

```python
iris.hist("petal_length",by="species")
```

## Boxplot charts

```python
iris.plot.box()
```

```python
colors={'boxes': 'Red', 'whiskers': 'blue','medians': 'Black', 'caps': 'Green'}
```

```python
iris.plot.box(color=colors)
```

```python
iris.plot.box(vert=False)
```

```python
iris.boxplot()
```

```python
plt.rcParams["figure.figsize"]=(8,8)
plt.style.use("ggplot")
iris.boxplot(by='species')
```
